### Evalucion de expresiones sobre un DataFrame

```python
# Crear un DataFrame e inicalizarlo con un diccioario de Objetos Series
Personas = {
"Peso": pd.Series([65,60,70,48,],["Mauricio", "Rebeca", "Arturo","Yulissa"]),
"Altura": pd.Series({"Mauricio":170,"Rebeca":160, "Arturo":169,"Yulissa":156}),
"Hijos": pd.Series([2,3],["Rebeca","Yulissa"])
}

df =pd.DataFrame(Personas)
  
df
```

```python
# Evaluar una funcion sobre una columna de DataFrame
df.eval("Altura / 2")
```

```python
# Asignar un valor resultante como una columna nueva
df.eval("Media_Altura = Altura / 2", inplace=True)
df
```

```python
# Evaluar una funcion utilizando una variable local
max_altura = 160 
df.eval("Altura > @max_altura")
```

```python
# Aplicar una funcion externa a una columna de DataFrame
def func(x):
    return x + 2

df["Peso"].apply(func)
```
